## Get Data from Wikipedia
###### Store and display HTML code from wikipedia link using BeatifulSoup.

```python
import requests
URL = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

from bs4 import BeautifulSoup
res = requests.get(URL).text
soup = BeautifulSoup(res,'lxml')
print(soup.prettify())
```

###### Explore data stored.

```python
print(soup.title)
print(soup.title.text)
print(soup.tr)
print(soup.tr.text)
print(soup.td)
```

###### Separate and display HTML code table.

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

###### Explore stored data.

```python
print(My_table.tr.text)
a = My_table.tr.text
a
```

###### Convert HTML code to string and then use display_html to view table.

```python
display_html(str(My_table), raw=True)
```

###### Convert HTML code to a dataframe.

```python
dfs = pd.read_html(str(My_table))
dfs
```

```python
df = dfs[0]
df
```

####### Remove rows in "Borough" column with string "Not assigned". Reset index of data frame.

```python
df = df[~df['Borough'].isin(['Not assigned'])]
df.reset_index(drop=True, inplace=True)
df
```

## Get Coordinates for Each Neighborhood

###### Import csv file from web.

```python
df_coor = pd.read_csv("https://cocl.us/Geospatial_data")
```

```python
df_coor.shape
```

###### Merge Toronto Neighborhood dataframe (df) and Neighborhood Coordinates dataframe (df_coor).

```python
df_com = pd.merge(df, df_coor, on='Postal Code')
df_com
```

## Explore and Cluster Neighborhoods in Toronto
###### Import libraries.

```python
import numpy as np # library to handle data in a vectorized manner

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

###### Use geopy library to get the latitude and longitude values of Toronto.

```python
address = 'Toronto, TO'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```

###### Create a map of Toronto with neighborhoods superimposed on top.

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_com['Latitude'], df_com['Longitude'], df_com['Borough'], df_com['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
